```python
def convert(obj):
  L=[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
def convert2(obj):
  M=[]
  j=0
  for i in ast.literal_eval(obj):
    if j<=3:
      M.append(i['name'])
    j+=1
  return M


```

```python
movies['cast']=movies['cast'].apply(convert2)
```

```python
def convert3(obj):
  N=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      N.append(i['name'])
  return N
```

```python
movies['crew']=movies['crew'].apply(convert3)
```

```python
movies=movies[['movie_id','title','overview','genres','keywords','release_date','cast','crew']]
```

```python
movies.head(1)
```

```python
movies['release_date'][0]
```

```python
def convert4(obj):
  l=[]
  l.append(obj[:4])
  return l
```

```python
movies['release_date']=movies['release_date'].apply(convert4)
```

```python
movies['release_date']
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head(1)
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
new_movie_df=movies[['movie_id','title','tags']]
```

```python
new_movie_df.head()
```

```python
new_movie_df['tags']=new_movie_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_movie_df.head()
```

```python
new_movie_df['tags'][0]
```

```python
new_movie_df['tags']=new_movie_df['tags'].apply(lambda x:x.lower())
```

```python
new_movie_df['tags'][1]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_movie_df['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y=[]

  for i in text.split():
    y.append(ps.stem(i))

    return " ".join(y)
```

```python
new_movie_df['tags']=new_movie_df['tags'].apply(stem)
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
  movie_index=new_movie_df[new_movie_df['title']==movie].index[0]
  distances=similarity[movie_index]
  movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
  for i in movie_list:
    print(new_movie_df.iloc[i[0]].title)
```
